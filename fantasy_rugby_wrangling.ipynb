```python
import pandas as pd
import numpy as np
```

```python
players_master = pd.read_csv("run_results.csv")
players_master['player_position'].fillna('x - Unknown', inplace=True)
players_master.columns = ['player', 'cost', 'country', 'position', 'tackles', 'd_tackles', 't_breaks', 'm_ball', 'mom', 'tries', 'conversions', 'penalties', 'd_goals', 'yellow', 'red', 'm_played']
players_master.set_index('player', inplace=True)
players_master.fillna(0, inplace=True)
players_master.replace({'14:15': 0, '16:45': 0, 'Deferred': 0}, inplace=True)
players_master['position'] = players_master['position'].apply(lambda x: x.split(' - ')[1])
players_master['position'] = players_master['position'].replace({'Fullback': 'Full Back', 'Scrum-half': 'Scrum Half', 'Back-row': 'Back Row', 'Fly-half': 'Fly Half', 'Second-row': 'Second Row'})
players_master[['d_tackles', 't_breaks', 'mom']] = players_master[['d_tackles', 't_breaks', 'mom']].astype('float')
players_master.sort_index(inplace=True)
players_master.head()
```

```python
players_training_week1 = pd.read_csv("players_training_week1.csv", index_col='player')

players_training_week1 = pd.merge(
    players_master[['country', 'position']],
    players_training_week1[['cost', 'mins', 'tackles', 'd_tackles', 't_breaks', 'm_ball', 'mom', 'tries', 'conversions', 'penalties', 'd_goals', 'yellow', 'red', 'started', 'week', 'm_played']],
    left_index=True,
    right_index=True,
    how='left'
)

players_training_week1.insert(0, 'cost', players_training_week1.pop('cost'))
players_training_week1.fillna(0, inplace=True)
players_training_week1.head()
```

```python
players_training_week2 = players_master
players_training_week2['week'] = 2
players_training_week2.head()
```

```python
players_test_week2 = pd.read_csv("players_test_week2.csv", index_col='player')
players_test_week2.head()
```

```python
players_training_week2 = pd.merge(players_training_week2, players_test_week2['started'], left_index=True, right_index=True, how='left')
```

```python
# players_training_week2['mins']
players_training_week2['tackles'] = players_training_week2['tackles'] - players_training_week1['tackles']
players_training_week2['d_tackles'] = players_training_week2['d_tackles'] - players_training_week1['d_tackles']
players_training_week2['t_breaks'] = players_training_week2['t_breaks'] - players_training_week1['t_breaks']
players_training_week2['m_ball'] = players_training_week2['m_ball'] - players_training_week1['m_ball']
players_training_week2['mom'] = players_training_week2['mom'] - players_training_week1['mom']
players_training_week2['tries'] = players_training_week2['tries'] - players_training_week1['tries']
players_training_week2['conversions'] = players_training_week2['penalties'] - players_training_week1['penalties']
players_training_week2['d_goals'] = players_training_week2['d_goals'] - players_training_week1['d_goals']
players_training_week2['yellow'] = players_training_week2['yellow'] - players_training_week1['yellow']
players_training_week2['red'] = players_training_week2['red'] - players_training_week1['red']
players_training_week2['m_played'] = players_training_week2['m_played'] - players_training_week1['m_played']

players_training_week2.head()
```

```python
players_training_week2.dropna(subset=['started'], inplace=True)
```

```python
def mins_estimator(started):
    """
    
    """
    
    df = players_training_week1
    
    if started == 1:
        return df[(df['started'] == 1)]['mins'].mean()
    else:
        return df[(df['started'] == 0) & (df['mins'] != 0)]['mins'].mean()
```

```python
players_training_week2['mins'] = players_training_week2['started'].apply(lambda x: mins_estimator(x))
```

```python
players_training_week2 = players_training_week2[['cost', 'country', 'position', 'mins', 'tackles', 'd_tackles',
       't_breaks', 'm_ball', 'mom', 'tries', 'conversions', 'penalties',
       'd_goals', 'yellow', 'red', 'started', 'week', 'm_played']]
```
